```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticks
import matplotlib.dates as mdates
plt.rcParams['figure.figsize'] = (12,3)

url ="https://docs.google.com/spreadsheets/d/e/2PACX-1vQN3Ksa9szQuO4G1-msXWAp17KtVHQCBnuEieu_auu1wSiBf3-krHusIx5VBMkihxj-KZLBosDIGEyR/pub?output=csv"
df = pd.read_csv(url)
df
```

```python
def summarize_df(df_):
    df = df_.copy()
    report = pd.DataFrame({
                        'dtype':df.dtypes,
                        '# null': df.isna().sum(),
                        'null (%)': df.isna().sum()/len(df)*100,
                        'nunique':df.nunique(),
                        "min":df.min(),
                        'max':df.max()
             })
    report.index.name='Column'
    return report.reset_index()
summarize_df(df)
```

# Part 1

```python
# Filter the 4 largest cities into a new dataframe.
# Tip: the "SizeRank" column has already ranked the cities by size. The larger the city, the smaller the rank value.
# Therefore the 4 largest cities would have rank values of [0,1,2,3]
```

```python
# Filter the 4 largest cities into a new dataframe.
big_city = df.loc[:3]
big_city
```

```python
#melt the dataframe to convert date features to rows
big_city_melt = pd.melt(big_city.drop(columns=['SizeRank']), 
                        id_vars=['RegionID', 'RegionName', 'RegionType', 'StateName', 'State', 'Metro', 'CountyName'],
                                     value_name='Price', var_name='Date')
big_city_melt
```

```python
#Convert the Date feature to the datetime format
big_city_melt['Date'] = pd.to_datetime(big_city_melt['Date'])
big_city_melt.info()
```

```python
# Convert the datetime feature into the index

big_city_melt = big_city_melt.set_index('Date')
big_city_melt
```

```python
#Resample the dataframe as monthly frequency, grouped by City.

big_city_res = big_city_melt.groupby('RegionName').resample('M').sum()
big_city_res
```

```python
def millions(x,pos):
    """function for use wth matplotlib FuncFormatter -  formats money in millions"""
    new_x = x/1_000_000
    return f"{new_x:,.0f}M"
def thousands(x,pos):
    """function for use wth matplotlib FuncFormatter -  formats money in millions"""
    new_x = x/1_000
    return f"{new_x:,.0f}K"
# Create the formatters
fmt_mill = mticks.FuncFormatter(millions)
fmt_thous = mticks.FuncFormatter(thousands)
```

```python
def plot_dates(ts,figsize=(16,4),xtick_fontsize='large', xtick_fontweight='bold',
               title= None):## Make the fig and axis first

    if title is None:
        title = ts.name
    # plot the time series with pandas
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot(ts.index, ts.values, marker='.', lw=1);
    #max_temp_yr.plot(ax=ax,title=title, style='.-');
   
    # create formatting to show prices in Thousands
    ax.yaxis.set_major_formatter(fmt_thous)
    
    ## creating our tick locators and formatters
    ## for minor month ticks
    # create the locator to place ticks every year.
    loc_3months = mdates.MonthLocator(interval=3)
    fmt_months = mdates.DateFormatter("%b")
    ## for major year ticks
    loc_year = mdates.YearLocator()
    fmt_year = mdates.DateFormatter("%Y")
    
    ## Change the font of the major ticks to stand out
#     ax.set_xticks(ax.get_xticks())
    ax.set_xticklabels(ax.get_xticklabels(), 
                       fontsize=xtick_fontsize, 
                       fontweight=xtick_fontweight) 
    
    
    ## customize minor ticks
    ax.xaxis.set_minor_locator(loc_3months)
    ax.xaxis.set_minor_formatter(fmt_months)
    ## customize major ticks
    ax.xaxis.set_major_locator(loc_year)
    ax.xaxis.set_major_formatter(fmt_year)
    ## Making major/minor gridlines visually distinct
    ax.grid(which='minor',axis='x',ls=":")
    ax.grid(which='major',axis='x',color='k')
    
     ## Change the font of the major ticks to stand out
#     ax.set_xticks(ax.get_xticks())
    ax.set_xticklabels(ax.get_xticklabels(), 
                       fontsize=xtick_fontsize, 
                       fontweight=xtick_fontweight) 

    ## rotate the dates
    fig.autofmt_xdate(which='major',rotation=90,ha='center')
    fig.autofmt_xdate(which='minor',rotation=90,ha='center')
    return fig,ax

plot_dates(city_unstacked)
```
